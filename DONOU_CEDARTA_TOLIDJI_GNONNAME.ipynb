***Question 3:  Dterminez les diffrentes strates ou aspects de ces images puis construisez un ensemble
de validation quilibr entre ces dernires en vous basant sur les infos de ce dataset
mentionn  la fin du notebook lesson1_cnn. Le ratio de lensemble de validation doit
tre de 25%.***

```python
classe = df.classname.value_counts()
fig = classe.plot(kind='bar')
```

```python
subject = df.subject.value_counts()
fig = subject.plot(kind='bar')
```

#### Cration du jeu de validation

```python
df = df.assign(for_test=False)

# On garde la pourcentage (pourcentage) de sujets par classe
pct = df.assign(nb=1).groupby(['subject', 'classname']).sum()
pct = pct.assign(pct = (pct.nb/df.shape[0]*100))
pct = pct.assign(subject=pct.index.get_level_values('subject'), classname=pct.index.get_level_values('classname'))
pct.index = range(0, pct.shape[0])


# pourcentage de chaque classe dans notre jeu de donnees
pct_classe = df.classname.value_counts() / df.shape[0] * 100


# on cre un dictionnaire contenant pour chaque classe et pour notre chantillon de test:
#    nb: pourcentage d'individu pour la classe qu'on aura  chaque itration, ,
#    max : correspond  la pourcentage de la classe dans notre jeu de donnes (dont on ne devra pas trop s'loigner),
#    subject: sujet slectionn pour la classe.
dico = dict()
for c in df.classname.unique():
    dico[c] = { 'nb': 0, 'max': pct_classe[c], 'subject': [] }
dico['t'] = { 'nb': 0, 'max': 25, 'subject': [] }


for c in df.classname.unique():
    # Les instruction  l'intrieur de cette boucle permettront de dterminer si le sujet sera pris en test ou pas
    for sub in pct.subject.unique():
        # la condition ci-dessous permet de vrifier pour notre chantillon de test si nous sommes en dessous des 25%,
        # et que pour chaque classe nousavons le pourcentage qui est respect dans le test.
        if sub not in dico['t']['subject'] and dico['t']['nb'] < 25 and sub not in dico[c]['subject'] and dico[c]['nb'] < dico[c]['max']:
            # cette boucle permet d'ajouter le sujet dans la partie subject de chaque classe vu que le sujet respecte les conditions prcdentes
            for c2 in df.classname.unique():
                dico[c2]['subject'].append(sub)
                dico[c2]['nb'] += sum(pct.loc[((pct.subject == sub) & (pct.classname == c2)), 'pct'])
           
            # on ajoute le sujet  la partie sujet de notre partie test
            dico['t']['subject'].append(sub)
            dico['t']['nb'] += sum(pct.loc[pct.subject == sub, 'pct'])


df.loc[df.subject.isin(dico['t']['subject']), 'for_test'] = True
```

```python
df.loc[:, "path"] = df.loc[:, "classname"].astype("str") + '/' + df.loc[:, "img"].astype("str")
```

```python
dls = ImageDataLoaders.from_df(
    df,
    'imgs/', 
    folder='train', 
    valid_col='for_test', 
    fn_col=4, 
    item_tfms=Resize(224),
    label_col=1,
    seed=seed)
```

```python
dls.show_batch()
```

***Question 4. Entranez votre modle (apprentissage par transfert) pendant 5 poques en prenant le
soin dutiliser le rythme dapprentissage optimal fourni par fastai.***

```python
learn = cnn_learner(dls, resnet18, metrics=[error_rate, accuracy])
```

Recherche du meilleur learning rate

```python
lrs = learn.lr_find(suggest_funcs=(minimum, steep, valley, slide))
```

```python
learn.fine_tune(epochs=5, base_lr=lrs.minimum, cbs=[ShowGraphCallback(),])
```

```python
learn.save("./somnolence_model_q4")
```

***(6. Entranez de nouveau votre modle en effectuant des changements de sorte  amliorer
la performance du modle. Il est recommand de lire la documentation afin de dcouvrir
certaines techniques dapprentissage amliorant la performance de votre modle.)***

```python
amelioration = [
        Rotate(360, p=1, mode='bilinear'), 
        Brightness(max_lighting=0.9, p=0.35),
        Contrast(max_lighting=0.4, p=0.25),
        RandomErasing(p=0.2, sl=0.0, sh=0.2, min_aspect=0.3, max_count=2),
        Flip(p=1),
        Zoom(max_zoom=1.1,p=0.1),
        RandomResizedCrop(224)
    ]


dls_ameliore = ImageDataLoaders.from_df(
    df,
    'imgs/', 
    folder='train', 
    valid_col='for_test', 
    fn_col=4, 
    item_tfms=Resize(224),
    batch_tfms=amelioration,
    label_col=1,
    seed=seed)

dls_ameliore.show_batch()
```

```python
learn_ameliore = cnn_learner(dls_ameliore, resnet18, metrics=[error_rate])
lrs_ameliore = learn_ameliore.lr_find(suggest_funcs=(minimum, steep, valley, slide))
```

```python
learn_ameliore.fine_tune(epochs=25, base_lr=lrs_ameliore.minimum, cbs=[ShowGraphCallback(),])
```

```python
learn_ameliore.save("./somnolence_model_q6")
```
